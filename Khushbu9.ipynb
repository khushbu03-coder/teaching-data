```python
import pandas as pd

# Create sample males_df
data_males = {'Waist_Circumference': [80, 85, 90],
              'Height': [170, 175, 180],
              'Hip_Circumference': [95, 100, 105]}
males_df = pd.DataFrame(data_males)

# Create sample females_df
data_females = {'Waist_Circumference': [70, 75, 80],
                'Height': [160, 165, 170],
                'Hip_Circumference': [90, 95, 100]}
females_df = pd.DataFrame(data_females)

# Compute ratios for males
males_df['Waist_to_Height_Ratio'] = males_df['Waist_Circumference'] / males_df['Height']
males_df['Waist_to_Hip_Ratio'] = males_df['Waist_Circumference'] / males_df['Hip_Circumference']

# Compute ratios for females
females_df['Waist_to_Height_Ratio'] = females_df['Waist_Circumference'] / females_df['Height']
females_df['Waist_to_Hip_Ratio'] = females_df['Waist_Circumference'] / females_df['Hip_Circumference']

# Display the updated dataframes
print("Males DataFrame with Ratios:")
display(males_df.head())

print("\nFemales DataFrame with Ratios:")
display(females_df.head())
```
