# Task
Create a new matrix zfemale being a version of the female dataset with all its columns standardised (by computing the z-scores of each column).

## Generate sample data

### Subtask:
Create a sample `female` dataset with numerical columns.


**Reasoning**:
The previous command failed because pandas was not imported. Import pandas and then create the DataFrame.



```python
import pandas as pd

data = {'Height': [160, 165, 170, 155, 162],
        'Weight': [55, 60, 65, 50, 58],
        'Age': [25, 30, 35, 22, 28],
        'BMI': [21.5, 22.0, 22.5, 20.8, 22.1]}
female = pd.DataFrame(data)
display(female)
```

## Standardize data

### Subtask:
Compute the mean and standard deviation for each column in the `female` dataset and calculate the z-scores to create the `zfemale` matrix.


**Reasoning**:
Compute the mean and standard deviation for each column in the `female` DataFrame and calculate the z-scores to create the `zfemale` matrix.



```python
female_means = female.mean()
female_stds = female.std()

zfemale = (female - female_means) / female_stds

print(zfemale.shape)
print("\nStandardized Female Data (First 5 rows):")
print(zfemale.head())
```

## Display results

### Subtask:
Show the shape and a preview of the `zfemale` matrix.


**Reasoning**:
Print the shape and the first 5 rows of the zfemale DataFrame to complete the subtask.



```python
female_means = female.mean()
female_stds = female.std()

zfemale = (female - female_means) / female_stds
```
